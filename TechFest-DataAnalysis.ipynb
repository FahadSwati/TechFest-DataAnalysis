## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import random
```

## CSV File
In this code, first I took data from TechFest Pakistan, which included 50+ participants, 12+ cities, 12+ jobs, and 12+ workshops. I converted that data into dictionaries, then created a DataFrame using Pandas, and finally converted it into a CSV file and imported it here.<br>
We will take data from the CSV file and compare the following fields:<br>
Name, City, Profession, Age, Workshop, Presentation_Score, and Feedback_Rating.<br>

```python
df=pd.read_csv("TechFest2025_100Participants.csv")  

df
```

## MEAN
It is for the average score.<br>
It adds all scores and divides by the number of participants.<br>
It also hows the overall performance level.<br>
its work like (Mean= sum of all scores divided by total number of participants)<br>
Helps to identify whether the group performed well or poorly on average.<br>
## MEDIAN
The median is the middle score when all values are arranged in ascending or descending order.<br>
If there is an odd number of scores, the middle value is the median.<br>
If there is an even number, the average of the two middle values is taken.<br>
It divides the dataset into two equal halves.<br>
## MODE
It is Most Frequent Value<br>
what score appeared most often.<br> 
Helps to identify what score or outcome is most common among participants.<br>
Useful in understanding dominant behavior or trends in the data.<br>

```python
print("Mean:", df['Presentation_Score'].mean())
print("Median:", df['Presentation_Score'].median())
print("Mode:", df['Presentation_Score'].mode()[0])

```

## Quartiles
Quartiles are essential in descriptive statistics to understand how our data is distributed.<br> They divide our dataset into four equal parts, helping us analyze it more precisely and meaningfully than just using the average (mean).<br>
## 25 Percentile (Q1)
Q1 (First quartile & 25 percentile)<br>
It is to understand 25% of the data falls below the given score.<br>
Also helps to understand lower range of performance.<br>
## 50 Percentile (Q2)
Q2 (second quartile & 50 percentile) also known as (Median) that shows the Middle Value<br>
It is to understand 50% of the data falls below the given score.<br>
Also helps to understand mid range of performance.<br>
## 75 Percentile (Q3)
Q3 (third quartile & 75 percentile).<br>
To understand 75% of the data falls below the given score.<br>
Also helps to understand uper range of performance.<br>
## Range
Range = Maximum - Minimum score.so,<br>
It tells us how spread out the scores are.<br>
It provides a quick summary of variability<br>
Itâ€™s the simplest measure of dispersion. Just subtract the smallest value from the largest.<br>

```python
print("25th Percentile:", np.percentile(df['Presentation_Score'], 25))
print("50th Percentile:", np.percentile(df['Presentation_Score'], 50))
print("75th Percentile:", np.percentile(df['Presentation_Score'], 75))
print("Range:", df['Presentation_Score'].max() - df['Presentation_Score'].min()) 
```

## Standard Deviation
There we use Deviation to  tells how much scores spread out from the mean or,<br>
How much the value closer to eachother.<br>
## Variance
It is the is the square of standard deviation.<br>
it is also a useful command. it shows the how much spread the value is.<br>
## Interquartile Range
it measures the spread of the middle 50% of  data.<br>
It is value obtain from difference between the First and Third quartile<br>
by this we'll find half of the quartile;s value<br>

```python
print("Std Deviation:", df['Presentation_Score'].std())    
print("Variance:", df['Presentation_Score'].var()) 
print("Interquartile Range (IQR):", np.percentile(df['Presentation_Score'], 75) - np.percentile(df['Presentation_Score'], 25))
```

## Histogram Bars
x-axis have scors<br>
y-axis have counts<br>
This graph shows the distribution of presentation scores for participants.<br>
Each bar represents how many participants scored within a certain score range.<br>
The smooth curve on top (in blue) shows the overall shape of the score distribution.<br>
It helps us understand whether the data is normally distributed, skewed, or has multiple peaks.<br>

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Presentation_Score'], kde=True, color='skyblue')
plt.title('Distribution of Presentation Scores')
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

## Boxplot graph
It is used for visually summarizing the distribution of a dataset.<br>
The box represents the interquartile range (IQR) â€“ the middle 50% of scores.<br>
Bottom of the box (Q1): 25th percentile<br>
Top of the box (Q3): 75th percentile<br>
The horizontal line inside the box is the median (50th percentile).<br>
The  lines extending above and below the box represent the range of the data called "whiskers".<br>

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Presentation_Score'], color="orange")
plt.title('Boxplot of Presentation Scores')
plt.ylabel("Score")
plt.show()
```

## Horizontal bar graph
Y-axis (City): Names of the cities from which participants came (e.g., Lahore, Karachi, Islamabad).<br>
X-axis (Count): Number of participants from each city.<br>
Bars: Represent how many participants are from each city.<br>
hue='City': Gives each bar a distinct color based on the city.<br>
palette='Set2': Applies a visually appealing color scheme to the bars.<br>
legend=False: Hides the legend since the city names are already on the y-axis.<br>
order=...value_counts(): Sorts cities in order of highest to lowest participant count.<br>

```python
plt.figure(figsize=(10,6))
sns.countplot(y='City', data=df, order=df['City'].value_counts().index, hue='City', palette='Set2', legend=False)
plt.title("Participants from Different Cities")
plt.xlabel("Count")
plt.ylabel("City")
plt.show()
```

## Horizontal bar chart
Y-axis: Different professions of the participants.<br>
X-axis: The number of participants belonging to each profession.<br>
Bars: Each bar represents how many participants belong to that specific profession.<br>
hue='Profession': Used to color each bar differently based on the profession.<br>
palette='Set2': This is a color palette that makes the bars visually distinct and easier to read.<br>
legend=False: The legend (color labels) is hidden to keep the graph clean, since the y-axis already shows the professions<br>
order=...value_counts(): Sorts Profession  in order of highest to lowest participant count.<br>

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Profession', data=df, order=df['Profession'].value_counts().index, hue = 'Profession', palette="Set2",legend=False)
plt.title("Number of Participants by Profession")
plt.xlabel("Count")
plt.ylabel("Profession")
plt.tight_layout()
plt.show()
```

## Horizontal bar graph
Y-axis shows the names of different workshops.<br>
X-axis shows the number of participants in each workshop.<br>
Each bar shows the total count of participants who joined that particular workshop.<br>
(Colors (Palette='Set3')): Different colors make the bars more visually clear and appealing.<br>
(Legend=False): The legend (color label guide) is turned off because each bar already clearly represents its own workshop.<br>
hue : it is used to seperate workshops by different colours.<br>
order=...value_counts(): Sorts workshops in order of highest to lowest participant count.<br>

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Workshop', data=df, order=df['Workshop'].value_counts().index, hue='Workshop', palette="Set3", legend=False)
plt.title("Workshop Participation")
plt.xlabel("Count")
plt.ylabel("Workshop")
plt.show()
```

## Scatter plot 
It shows the relationship between a participantâ€™s Age and their Presentation Score.<br>
It also shows participent's Feedback rating using color.<br>
It contain Age,Score,Feedback<br>
x-axis shows the sores of participants<br>
y-axis shows the age of participants<br>
hue is used for colour distribution for Feedbacck rating<br>
palette = coolwarm is used to difference between higher and lower rating<br>
red colour for higher rating and blue for lower as shown in graph.<br>

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Age', y='Presentation_Score', hue='Feedback_Rating', palette='coolwarm')
plt.title("Age vs Presentation Score")
plt.xlabel("Age")
plt.ylabel("Presentation Score")
plt.legend(title='Feedback Rating', loc='lower left',bbox_to_anchor=(1.05, 0))
plt.show()
```
