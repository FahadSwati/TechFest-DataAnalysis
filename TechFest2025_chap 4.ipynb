## Importing Libraries.
Load  TechFest dataset using pandas.<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing CSV.
Here the CSV file is imported to read data.<br>

```python
# Load the dataset
df = pd.read_csv('TechFest2025_participants.csv')

# Show basic info
df.info()
```

## Histogram.
Code creates a histogram to visualize the distribution of participant ages in TechFest2025 dataset..<br>
### Purpose. 
#### Part--------------------------------------Purpose
df['Age'].dropna()-------------------Removes any missing (NaN) age values.<br>
bins=15------------------------------Divides age values into 15 intervals (groups).<br>
kde=True-----------------------------Adds a smooth curve showing the probability density (distribution shape).<br>
color='skyblue'----------------------Sets a light blue color for the bars.<br>
plt.figure(figsize=(10, 6))----------Makes the plot wide and clear.<br>
### Findings.
Shows how participant ages are distributed:.<br>
For example, seen that most participants are aged between 25–40.<br>
Helps identify age groups that are most common.<br>
The KDE curve tells  whether the distribution is normal, skewed, or multimodal.<br>
This is useful for event planning — knowing which age group is most engaged helps in designing better content.<br>

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=15, kde=True, color='skyblue')
plt.title('Histogram of Participant Ages')
plt.xlabel('Age')
plt.ylabel('count')
plt.show()
```

## Boxplot.
Code creates a boxplot to analyze the distribution and spread of the Presentation_Score column.<br>
### Purpose.
#### Component-----------------------------Purpose
sns.boxplot(...)----------------Draws the boxplot horizontally for Presentation_Score.<br>
plt.figure(...)-----------------Adjusts the width of the chart.<br>
plt.title & plt.xlabel----------Adds labels and title for clarity.<br>
### Findings.
A boxplot gives summary statistics visually, including:.<br>
#### Feature-----------------------------What It Shows
Median (middle line)---------------The average score of participants.<br>
Box (IQR)--------------------------The range where 50% of scores lie (between Q1 and Q3).<br>
Whiskers---------------------------The full range of normal values.<br>
Dots (outliers)(If Any..)----------Any unusually high or low scores (participants who scored much higher or lower than others).(No outliers in data).<br>

### Insights.
If the box is tight, most participants scored similarly.<br>
If there are many dots (outliers), some participants scored much higher/lower than others.(No outliers in data).<br>
If the box is skewed, the scores are not evenly distributed.<br>
This helps identify inconsistencies or performance trends in TechFest presentations.<br>

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Presentation_Score'])
plt.title('Boxplot of Presentation Scores')
plt.xlabel('Presentation Score')
plt.show()
```

## Boxplot.

### Purpose.

## Violin plot.
This code creates a violin plot to visualize the distribution and density of values in the Feedback_Rating column.<br>
It’s like a combination of a boxplot + distribution curve.<br>
### Purpose.
#### Component----------------------------Function
sns.violinplot(...)--------------Draws the violin plot for the Feedback_Rating column.<br>
plt.figure(...)------------------Sets the size of the plot.<br>
plt.title, plt.xlabel------------Adds title and axis label for clarity.<br>
### Findings.
Width of the violin at each point shows how many participants gave that rating.<br>
The fatter the area, the more participants gave that score.<br>
Includes a white dot (median) and a thin box showing the interquartile range (IQR).<br>

### Insights.
See which feedback scores were most common.<br>
It reveals if participants were generally satisfied (4–5) or critical (1–2).<br>
If the plot is symmetrical and centered high, feedback was generally very good.<br>
If there are multiple peaks, feedback opinions were diverse.<br>
This is super useful for understanding audience satisfaction with TechFest sessions or organization.<br>

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['Feedback_Rating'])
plt.title('Violin Plot of Feedback Ratings')
plt.xlabel('Feedback Rating')
plt.show()
```

## count plot.
This code creates a count plot (a bar chart) that displays the number of participants in each profession.<br>
It helps you understand the professional background of TechFest audience.<br>
### Purpose.
#### Part-------------------------What It Does
sns.countplot(...)----------Counts how many times each profession appears in the dataset.<br>
hue='Profession'------------Colors each bar separately (optional for styling).<br>
palette='Set2'--------------Applies soft, light colors to bars.<br>
legend=False----------------Hides the redundant legend since labels are already on the x-axis.<br>
rotation=45-----------------Rotates x-axis labels to avoid overlap.<br>
### Findings.
Shows which professions are most or least represented at TechFest.<br>
Helps to identify whether most participants are Developers, Students, Teachers, Researchers, etc.<br>
Useful for tailoring content — e.g., if most are students, make sessions beginner-friendly.<br>

### Insights.
If “Student” and “Developer” have the tallest bars: these are your main audience.<br>
If some bars are very short: those professions may need better outreach.<br>
Easy way to visually profile your participant community.<br>

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Profession', hue='Profession', palette='Set2', legend=False)
plt.title('Count of Participants by Profession')
plt.xlabel('Profession')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Bar chart.
This code creates a bar chart that shows how many participants belong to each specific age group.<br>
### Purpose.
#### Part-----------------------------------------Description
df['Age'].value_counts()-----------------Counts how many times each age appears in the dataset.<br>
plot(kind='bar')-------------------------Draws a vertical bar chart.<br>
color='teal'----------------------------Sets the bar color.<br>
figsize, title, xlabel, ylabel----------Customize chart size and labels.<br>
xticks(rotation=45)---------------------Rotates x-axis labels to avoid overlapping if there are many unique ages.<br>
### Findings.
Which ages are most common among TechFest participants.<br>
Whether there’s a dominant age group (e.g., 25, 30, 35).<br>
If participation is evenly spread or skewed toward certain ages.<br>

### Insights.
If one or two ages have much taller bars, it means most participants were that exact age.<br>
A wide spread of bars means TechFest attracted a diverse age range.<br>
Helpful for event targeting — for example, if most participants are in their 20s, you can customize content for young professionals.<br>

```python
plt.figure(figsize=(12, 6))
df['Age'].value_counts().plot(kind='bar', color='teal')
plt.title('Participants by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Pie chart.
This code draws a pie chart to show the percentage of participants from each city.<br>
### Purpose.
#### Part--------------------------------What It Does
df['City'].value_counts()--------Counts the number of participants from each city.<br>
plot.pie(...)--------------------Creates the pie chart.<br>
autopct='%1.1f%%'----------------Displays percentage values on the chart (1 decimal place).<br>
startangle=140-------------------Rotates the start of the pie chart for better orientation.<br>
plt.ylabel('')-------------------Removes unnecessary y-axis label for a cleaner look.<br>
### Findings.
Shows which cities most of the participants came from.<br>
Highlights geographical dominance — for example, if most attendees were from Karachi or Lahore.<br>
Easily shows proportional representation by city.<br>

### Insights.
Helps in regional planning: cities with higher percentages can be prioritized for future events.<br>
Cities with lower representation may need better marketing or outreach.<br>
Useful for generating regional performance or impact reports.<br>

```python
plt.figure(figsize=(6, 6))
df['City'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('City Distribution')
plt.ylabel('')
plt.show()
```

## Pie chart.
Code builds a pie chart that displays the percentage of participants who fall under each participation status category.<br>
### Purpose.
#### Code Part------------------------------------------------Role
df['Participation_Status'].value_counts()------------------Counts how many participants are in each status (e.g., Active, Dropped, Pending).<br>
.plot.pie(...)---------------------------------------------Converts the count into a percentage-based pie chart.<br>
autopct='%1.1f%%'------------------------------------------Displays exact percentages on the pie slices.<br>
startangle=140---------------------------------------------Rotates the pie for better readability.<br>
ylabel('')-------------------------------------------------Hides the y-axis label for a cleaner chart.<br>
### Findings.
Shows how many participants actively engaged, dropped out, or were pending.<br>
Makes it easy to compare proportions visually.<br>
Helps evaluate event retention or drop-off rates.<br>

### Insights.
If most participants are “Confirmed”, it shows strong event engagement.<br>
If many are “Canclled” or “Pending”, it may indicate issues like poor follow-up or low interest.<br>
Helps the event team make decisions about follow-ups, rewards, or improvements.<br>

```python
plt.figure(figsize=(6, 6))
df['Participation_Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Participation_Status')
plt.ylabel('')
plt.show()
```
